```python
import numpy as np
import pandas as pd

data = [
    [101, "Arun", "Developer", 2, 25, 8000],
    [102, "Bala", None, 5, 30, 12000],
    [103, "Charan", "Analyst", 3, 28, 9500],
    [104, "Divya", "Manager", 8, 40, 15000],
    [105, "Esha", None, 1, 22, 7000]
]

columns = ["empid", "name", "designation", "years_of_experience", "age", "salary"]
df = pd.DataFrame(data, columns=columns)

print("Original DataFrame:\n")
print(df)

df["age"] = df["age"] + 3

df.rename(columns={"years_of_experience": "exp"}, inplace=True)

df = df[df["salary"] <= 10000]

df["designation"] = df["designation"].fillna("Not Assigned")

print("\nUpdated DataFrame:\n")
print(df)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob'],
    'Age': [25, 30]
})

df2 = pd.DataFrame({
    'Name': ['David', 'Eva'],
    'Age': [35, 28]
})

print("DataFrame 1:")
print(df1)

print("\nDataFrame 2:")
print(df2)

df_concat_col = pd.concat([df1, df2], axis=1)
print("\nConcatenate Column-wise:")
print(df_concat_col)

df_concat_row = pd.concat([df1, df2], axis=0)
print("\nConcatenate Row-wise:")
print(df_concat_row)

df3 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Eva'],
    'Marks': [85, 90, 88]
})

df_merge = pd.merge(df1, df3, on='Name', how='inner')
print("\nMerged DataFrame:")
print(df_merge)
```

```python
import numpy as np

arr1 = np.array([[25, 30]])
arr2 = np.array([[35, 28]])

print("Array 1:")
print(arr1)

print("\nArray 2:")
print(arr2)

concat_col = np.concatenate((arr1, arr2), axis=1)
print("\nConcatenate Column-wise:")
print(concat_col)

concat_row = np.concatenate((arr1, arr2), axis=0)
print("\nConcatenate Row-wise:")
print(concat_row)

print("\nUsing hstack:")
print(np.hstack((arr1, arr2)))

print("\nUsing vstack:")
print(np.vstack((arr1, arr2)))
```
